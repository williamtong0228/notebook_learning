```python
from PIL import Image
```

```python
def plt_to_np():
    fig = plt.figure()
    fig.add_subplot(111)
    fig.tight_layout(pad=0)
    fig.canvas.draw()
    data = np.fromstring(fig.canvas.tostring_rgb(), dtype=np.uint8, sep='')
    data = data.reshape(fig.canvas.get_width_height()[::-1] + (3,))
    return data
```

```python
def drop_res(x,scale):
    size = (x.shape[0], x.shape[1])
    shrunk_size = (int(size[0]/scale),int(size[1]/scale))
    
    
    shrunk_img = np.array(Image.fromarray(x).resize(shrunk_size),Image.NEAREST) 
    lr_img = np.array(Image.fromarray(shrunk_img).resize(size),Image.NEAREST) 
    
    return lr_img


```

```python
np_image = plt_to_np()
np_image
```

```python
lr_img = drop_res(np_image, scale)
```
